```python
## Document를 Matrix로 변환하는 함수
def doc2mat(doc): #stem 된것만 넣기
    result = []
    for i in doc.split():
        if PorterStemmer().stem(i) not in swstemmed:
            if i in source_sg_model:
                v = source_sg_model[i]
                result.append(v)
    return np.matrix(result)
```

```python
## Document를 Matrix로 변환하는 함수
def doc2matnorm(doc): #stem 된것만 넣기
    result = []
    for i in doc.split():
        if PorterStemmer().stem(i) not in swstemmed:
            if i in source_sg_model:
                v = source_sg_model[i]
                result.append(v/np.linalg.norm(v))
    return np.matrix(result)
```

```python
##'Euclidean distance score method'
def ed(a, b):
    if a not in source_sg_model or b not in source_sg_model:
        return 0
    else:
        return np.exp(-(np.linalg.norm(source_sg_model[a]-source_sg_model[b]))/15.516)
```

```python
'1: cosine 유사도 계산'
import numpy as np

def similarity(a, b):
    if a not in source_sg_model or b not in source_sg_model:
        return 0
    else:
        av = source_sg_model[a]
        bv = source_sg_model[b]
        n1 = np.linalg.norm(av)
        n2 = np.linalg.norm(bv)
        return np.dot(av, bv) / n1 / n2
```

```python
##np.exp(-(np.linalg.norm(source_sg_model[a]-source_sg_model[b])))
##Documnet와 주어진 string간의 점수 비교
##Testmode: 코드 오류 검사시 True로 설정
def documentscoremated(s, doc, window = 3, power = 1000, scoretype = 'ed', testmode = False, threshold = 0.8, shortwordidx = []):#d: docmat, s: query mat, window<bound여야함!
    th = threshold
    totalscore = 0
    score = 0
    nwords = 0
    n = 0
    idx = 0
    L = len(doc)
    if testmode == True:
        print('L:',L)
    if scoretype == 'ed':
        scoremat = np.matrix([[]]*L)
        for i in s:
            mat = doc - i
            #print(len(mat))
            scoremat = np.append(scoremat, np.transpose(np.matrix(np.exp(-(np.linalg.norm(mat, axis = 1))/15.516))), axis = 1)
        if testmode == True:
            print('scoremat',scoremat)
            print('doc',doc)
            print('s',s)
            print('mat',mat)
            print('norm',np.linalg.norm(mat, axis =1))
    elif scoretype == 'cs':
        scoremat = doc * np.transpose(s)
        if testmode == True:
            print('scoremat',scoremat)
            print('doc',doc)
            print('s',s)
    qlen = len(s)
    qsize = range(qlen)
    bound = L - window
    for i in range(window):
        lb = 0
        ub = i + window + 1
        
        for k in qsize:    
            if k in shortwordidx:
                threshold = 0.96
            else:
                threshold = th
            if scoremat[i, k] > threshold:
                dfraction = range(lb, ub)
                nwordlist = []
                for j in dfraction:
                    for l in qsize:
                        if l in shortwordidx:
                            threshold = 0.96
                        else:
                            threshold = th
                        a = scoremat[j, l]
                        if a > threshold and l not in nwordlist:
                            score += a #벡터곱으로
                            nwordlist.append(l)
                            break
                nwords = len(nwordlist)
                if testmode == True:
                    print(scoremat[i, k])
                    print(i,k,nwords)
                totalscore += (score/nwords**2)*power**(nwords)
                score = 0
                break
    for i in range(window, bound):
        
        lb = i - window
        ub = i + window + 1
        
        for k in qsize:   
            if k in shortwordidx:
                threshold = 0.96
            else:
                threshold = th
            if scoremat[i, k] > threshold:
                dfraction = range(lb, ub)
                nwordlist = []
                for j in dfraction:
                    for l in qsize:
                        if l in shortwordidx:
                            threshold = 0.96
                        else:
                            threshold = th
                        a = scoremat[j, l]
                        if a > threshold and l not in nwordlist:
                            score += a #벡터곱으로
                            nwordlist.append(l)
                            break
                nwords = len(nwordlist)
                if testmode == True:
                    print(scoremat[i, k])
                    print(i,k,nwords)
                totalscore += (score/nwords**2)*power**(nwords)
                #print(score, totalscore)
                score = 0
                break
    for i in range(bound, L):    

        lb = i - window
        ub = L

        for k in qsize:
            if k in shortwordidx:
                threshold = 0.96
            else:
                threshold = th
            if scoremat[i, k] > threshold:
                #i*벡터곱(cosinesimilarity) > xx이상
                dfraction = range(lb, ub)
                nwordlist = []
                for j in dfraction:
                    for l in qsize:
                        if l in shortwordidx:
                            threshold = 0.96
                        else:
                            threshold = th
                        a = scoremat[j, l]
                        if a > threshold and l not in nwordlist:
                            score += a #벡터곱으로
                            nwordlist.append(l)
                            break
                nwords = len(nwordlist)
                if testmode == True:                    
                    print(scoremat[i, k])
                    print(i,k,nwords)
                totalscore += (score/nwords**2)*power**(nwords)
                #print(score, totalscore)
                score = 0
                break
    return totalscore
```

```python
##문서내 단어 빈도를 dictionary로 나타내는 함수
def wordfreq(txt):
    txtlist = txt.split()
    d = {}
    for w in txtlist:
        if w in d:
            d[w] += 1
        else:
            d[w] = 1
     
    return d
```

```python
##백업
##교과서 검색기
## type= document인 경우 p가 적용됨.
def textsearchbysectionmatedo(s, window = 3, kw = 10, type = 'simple', p = 10, translation = True): #doc와 query 둘다 stemm완료
    w = window
    start = time.time()
    category = []
    hlen = len(hcontentname)
    nslen = len(nscontentname)
    cnameall = hcontentname + nscontentname
    if type == 'simple':
        if translation == True or translation == 'quick': #번역-한글 tokenization은 시행x, 번역 후 1단어 이상일때 tokenization은 시행
            s = hsmanager(translator(s))
        shortwords = []
        ssplit = s.split()
        for i in range(len(ssplit)): #짧은단어의 경우 threshold를 높게잡기
            if len(ssplit[i]) <= 4:
                shortwords.append(i)
        #print('shortwords', shortwords)
        for i in range(hlen):
            score = documentscorematedo(doc2mat(hsmanager(s)), Hsectionlistmat[i], window = w, shortwordidx = shortwords)
            category.append(score)
        for i in range(nslen):
            score = documentscorematedo(doc2mat(hsmanager(s)), NSsectionlistmat[i], window = w, shortwordidx = shortwords)
            category.append(score)
        categorysorted = sorted(category, reverse=True)
        for i in range(len(categorysorted)): #score를 순위별로 정렬하고 결과를 나타냄
            if categorysorted[i] < categorysorted[0]*0.3:
                printindex = i
                break
    elif type == 'document':
        file = open('./Documents/'+s+'.txt', 'r', encoding='utf-8').read()
        freqdictionary = wordfreq(hsmanager(file))
        freqdictionarytr = {}
        if translation == 'quick': #빈도상위 몇단어만 꼽아서 번역
            sortedbyfreqquick = sorted(freqdictionary.items(), key=lambda word: word[1], reverse = True)
            freqdictionary = {k[0]:k[1] for k in sortedbyfreqquick[:kw*2]}
            #print(freqdictionary)
        if translation == True or translation == 'quick': #번역-한글 tokenization은 시행x, 번역 후 1단어 이상일때 tokenization은 시행
            for i in freqdictionary:
                similarpresence = False
                if langdetector(i) == 'ko':
                    itranslated = hsmanager(translator(i))
                    itranslatedspl = itranslated.split() #translation한 후 단어가 두개일 경우 list로 변환
                    for j in itranslatedspl:
                        t = 0.9
                        if len(j) <= 4:
                            t = 0.96 #짧은단어는 거의 같은 단어만 유사하다고 인정(threshold를 높임)
                        for k in freqdictionarytr:
                            if ed(j,k) > t:
                                freqdictionarytr[k] += freqdictionary[i]
                                similarpresence = True
                        if similarpresence == False:
                            freqdictionarytr[j] = freqdictionary[i]
                else:
                    t = 0.9
                    if len(i) <= 4:
                        t = 0.96
                    for k in freqdictionarytr:
                        if ed(i,k) > t:
                            freqdictionarytr[k] += freqdictionary[i]
                            similarpresence = True
                    if similarpresence == False:
                        freqdictionarytr[i] = freqdictionary[i]
            freqdictionary = freqdictionarytr
        sortedbyfreq = sorted(freqdictionary.items(), key=lambda word: word[1], reverse = True)
        keywords = [x[0] for x in sortedbyfreq if x[0] not in swstemmed and x[0].isnumeric() == False]
        if translation == 'deletion': #pseudo-document를 작성. translation이 deletion인 경우 top10중 한국어를 제외.
            pseudodoc = ' '.join([x for x in keywords[0:kw] if langdetector(x) != 'ko'])
        else:
            pseudodoc = ' '.join(keywords[0:kw])
        shortwords = []
        ssplit = pseudodoc.split()
        for i in range(len(ssplit)): #짧은 단어들의 경우를 저장
            if len(ssplit[i]) <= 4:
                shortwords.append(i)
        print('shortwords', shortwords)
        #print(freqdictionary)
        print(keywords[:kw])
        print(pseudodoc)
        for i in range(hlen): #각 단원별 score를 매김
            score = documentscorematedo(doc2mat(pseudodoc), Hsectionlistmat[i], window = w, power = p, shortwordidx = shortwords)
        #    for j in ssplit:
        #        score += documentscoremated(doc2mat(hsmanager(j)), Hsectionlistmat[i], w)
            category.append(score)
        for i in range(nslen):
            score = documentscorematedo(doc2mat(pseudodoc), NSsectionlistmat[i], window = w, power = p, shortwordidx = shortwords)
            #for j in ssplit:
            #    score += documentscoremated(doc2mat(hsmanager(j)), NSsectionlistmat[i], w)
            category.append(score)
        categorysorted = sorted(category, reverse=True)
        for i in range(len(categorysorted)): #score를 순위별로 정렬하고 결과를 나타냄
            if categorysorted[i] < categorysorted[0]*0.27:
                printindex = i
                break
    d = 0
    for i in range(printindex):
        if categorysorted[i] == categorysorted [i-1]:
            d += 1
            idx = [k for k, e in enumerate(category) if e == categorysorted[i]][d]##중복점수일경우
        else:
            d = 0
            idx = category.index(categorysorted[i])
        print('Recommendation-'+str(i+1)+': '+cnameall[idx])
        print('score:', categorysorted[i])
    print('elapsed time:', time.time()-start)
```
